```python
plt.figure(figsize=(15,5))
a=sns.countplot(x='Size',data=df,color='green')

```

```python
a=sns.countplot(x='Size',data=df,color='orange')
for bars in a.containers:
    a.bar_label(bars)
```

NOTE= FROM ABOVE GRAPH WE CAN SEE THAT MOSTLY OR MAXI. ORDER SIZE IS M

```python
df['Category']=df['Category'].astype(str)
```

```python
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=20,edgecolor='red')
plt.xticks(rotation=45)
plt.show()
```

NOTE= FROM ABOVE GRAPH WE CAN SEE MAXI. PRODUCT ORDER IS TSHIRT

```python
B2B_check=df['B2B'].value_counts()

plt.pie(B2B_check,labels=B2B_check,autopct='%1.1f%%')
plt.show()
```

```python
B2B_check=df['B2B'].value_counts()

plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

NOTE=FROM above chart we can see that maxi. 99.3% of buyers are retailers and 0.7% are B2b BUYERS

```python
x=df['Category']
y=df['Size']
plt.scatter(x,y)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter plot')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state',color='red')
plt.xlabel('Ship-Size')
plt.ylabel('Count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#top10 state
top_10_state=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state',color='red')
plt.xlabel('Ship-Size')
plt.ylabel('Count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```
